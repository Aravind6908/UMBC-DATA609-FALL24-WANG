```python
import pandas as pd
import plotly.express as px

pd.set_option("display.max_rows", None)

df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

df.shape
```

```python
# Getting duplicate rows if any
duplicate_rows = df3.duplicated()
duplicate_rows = duplicate_rows.reset_index().rename(columns={"index": "row_num", 0: "isDuplicate"})
# check for duplicate_rows
duplicate_rows[duplicate_rows['isDuplicate']==True]
```

## HW Question 2

Find out if there is a relationship between wealth and health

Pick any year worth of data and use plotly express to generate a scatter plot with GDP as x and life expectancy as y.

```python
correlation = df3['GDP per capita (current US$)'].corr(df3['Life expectancy at birth, total (years)'])
print("Correlation coefficient:", correlation)
```

The correlation analysis tells us how the wealth of a person is significant to the health of the person in that country.
From the above result we can see that the value is 0.59 which is greater than 0 which tells that the health is positively significant to the wealth which is that the more the wealth, more the health.

```python
df_2021 = df3[df3["Year"] == 2021].dropna()
df_2021.head()
```

```python
# Find out if there is a relationship between wealth and health

df_2021[['GDP per capita (current US$)', 'Life expectancy at birth, total (years)']].head()
```

```python
# Pick any year worth of data and use plotly express to generate a scatter plot with GDP as x and life expectancy as y.
fig = px.scatter(df_2021, x='GDP per capita (current US$)', y='Life expectancy at birth, total (years)',
                 title='Relationship between GDP per capita (Wealth) and Life Expectancy(Health)')

# Show the plot
fig.show()
```

From the above we can see that the life expectancy of the people whose GDP is more than 75k is always above 75. So there is a relation ship more the GDP (Wealth) more the life expectancy(Health)

## Homework Question 3

Select a year of data. calculate total population of each region and make a pie chart to display the distribution of total population of all regions.

```python
df_2021['Region'].value_counts()
```

```python
population_sum = df_2021.groupby(['Region']).agg({'Population, total': ['sum']}).reset_index()
population_sum.columns = ['_'.join(col) for col in population_sum.columns]
population_sum
```

```python
fig = px.pie(population_sum, values='Population, total_sum', names='Region_', title='Population by Region')
fig.show()
```

## Home work question 4

1. remove the color legend.
2. Order the above bar chart based on the values of the average GDP per capita.

```python
# Sort df_group by GDP per capita (current US$) in descending order
df_group = df_group.sort_values(by='GDP per capita (current US$)', ascending=False)

fig = px.bar(df_group, x="Region", y="GDP per capita (current US$)", color="Region",   text="GDP per capita (current US$)")
# Remove the legend
fig.update_layout(showlegend=False)

fig.show()
```
