## Step 4 - Categorize Each Country

Create a new column "Group". If a country is one of the seven G7 countried, the value of "Group" would be "G7", if a country of one of the five BRICS countries, the value of "Group" would be "BRICS". Otherwise, the value of "Group" would be "REST of WORLD".

```python
def groupie(x):
  if x in ["Canada", "France", "Germany", "Italy", "Japan", "United Kingdom", "United States"]:
    return "G7"
  elif x in ["Brazil", "Russia", "India", "China", "South Africa"]:
    return "BRICS"
  else:
    return "REST of WORLD"
```

```python
df['Group'] = df['Country Name'].apply(groupie)
df[['Country Name', 'Group']]
```

## Step 5 - Aggregate Data

Aggregate the following columns at the "Group" level for each year:

- For GDP, using "sum" for aggregation.
- For GDP Per Capita, use "mean" for aggregation.
- Check your calculation with the reference site.

```python
df_agg = df.groupby(['Year', 'Group']).agg({'GDP ($Billions)': 'sum', 'GDP Per Capita ($)': 'mean'})
df_agg
```

## Step 6 - Calculate Percentage Share of GDP

For each year, calculate the percentage share of each group's GDP. Check your calculation with the reference site.

```python
df_agg = df_agg.reset_index()
df_agg
```

```python
df_agg = df_agg.merge(df_agg.groupby('Year')[['GDP ($Billions)', 'GDP Per Capita ($)']].sum().reset_index(), on='Year')
df_agg
```

```python
df_agg['GDP Percentage Share'] = (df_agg['GDP ($Billions)_x'] / df_agg['GDP ($Billions)_y'])*100
df_agg['GDP Per Capita Percentage Share'] = (df_agg['GDP Per Capita ($)_x'] / df_agg['GDP Per Capita ($)_y'])*100
df_agg.head()
```

## Step 7 - Plot the GDP Percentage Share Over Time

```python
fig = px.line(df_agg, x='Year', y='GDP Percentage Share', color='Group',
              title="GDP Percentage Share Over Time",
              labels={'Percentage_Share': 'GDP Percentage Share', 'Year': 'Year'},
              markers=True)

fig.update_layout(yaxis_tickformat=".2f")  # Format y-axis to show 2 decimal places
fig.show()
```

From the above plot we can see that the GDP percentage of G7 countries have gone decreased over the last 10 -1 2 years where as countries BRICS have increased their GDP percentage  over the last 10 years.

## Step 10 - Plot the 2023 GDP Pie Charts

One pie chart for G7 and one pie chart for BRICS. Each country is a slice of a pie. Make sure the following information is displayed for each slice:

- Country Name
- GDP Percentage Share
- GDP Total

```python
df_2023  = df[df['Year']==2023]
def plot_pie_chart(df, group_name):
    # Filter data for the specified group
    df_group = df[df['Group'] == group_name]

    # Calculate total GDP for the group
    total_gdp = df_group['GDP ($Billions)'].sum()

    # Calculate percentage share for each country within the group
    df_group['Percentage_Share'] = (df_group['GDP ($Billions)'] / total_gdp) * 100

    # Create the pie chart
    fig = go.Figure(data=[go.Pie(
        labels=df_group['Country Name'],
        values=df_group['GDP ($Billions)'],
        hole=0.3,
        textinfo='label+percent',
        hovertemplate='<b>Country:</b> %{label}<br>' +
                      '<b>GDP Percentage Share:</b> %{percent}<br>' +
                      '<b>GDP Total:</b> %{value}<extra></extra>'
    )])

    # Update layout
    fig.update_layout(
        title=f"GDP Percentage Share by Country in {group_name}",
        annotations=[dict(text=group_name, x=0.5, y=0.5, font_size=20, showarrow=False)]
    )

    fig.show()

# Plotting pie charts for G7 and BRICS
plot_pie_chart(df_2023, 'G7')
plot_pie_chart(df_2023, 'BRICS')
```

From the avove donut charts for G7 and BRICS we can see that United states from G7 has the highest share and China fromt the BRICS share the highest GDP share among all the other countries.
