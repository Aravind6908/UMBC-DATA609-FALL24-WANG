```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Create the base line chart
fig = px.line(
    df_quad,
    x="Year",
    color="Country",
    y="GDP per capita (current US$)"
)

# Remove legend
fig.update_layout(showlegend=False)

# Add labels at the end of each line
for country in df_quad["Country"].unique():
    # Filter the data for each country
    df_country = df_quad[df_quad["Country"] == country]

    # Add the trace with labels at the last point
    fig.add_trace(
        go.Scatter(
            x=df_country["Year"],
            y=df_country["GDP per capita (current US$)"],
            mode="lines+text",
            name=country,
            text=[country if i == len(df_country) - 1 else "" for i in range(len(df_country))],
            textposition="middle right"
        )
    )

# Show the chart
fig.show()
```

## 10/09/2024 Homework Question 2

Round the Average GDP for each region to 2 decimal points.
Sort the bar according to GDP either in ascending order. Currently, it is ordered alphabetically.

```python
df_filtered = df[df["Year"] == 2010].dropna(subset=["Population, total", "GDP per capita (current US$)", "Region"])
print(df_filtered.shape)
df_filtered.sample(5)
```

```python
gdp_by_region = df_filtered.groupby('Region').agg(
    {
        'GDP per capita (current US$)': 'mean'
    }
).reset_index()

gdp_by_region
```

```python
fig = px.bar(
    data_frame=gdp_by_region,
    x='Region',
    color="Region",
    y='GDP per capita (current US$)',
    title='Average GDP per Capita by Region',
    labels={
        'GDP per capita (current US$)': 'Average GDP per Capita (Current US$)',
        'Region': 'Region'
    },
    text='GDP per capita (current US$)'
)


fig.show()
```

Rounding the GDP to decimals

```python
gdp_by_region['rounded_gdp'] = gdp_by_region['GDP per capita (current US$)'].round(2)
gdp_by_region.head()
```

```python
# Sort the data by 'GDP per capita (current US$)' in ascending order
gdp_by_region_sorted = gdp_by_region.sort_values(by='rounded_gdp', ascending=True)

# Create the bar chart with sorted data
fig = px.bar(
    data_frame=gdp_by_region_sorted,
    x='Region',
    color='Region',
    y='rounded_gdp',
    title='Average GDP per Capita by Region',
    labels={
        'Average GDP per Capita by Region': 'Average GDP per Capita by Region',
        'Region': 'Region'
    },
    text='rounded_gdp'  # Add text labels on bars
)

# Display the chart
fig.show()
```

## 10/09/2024 Homework question 3

Group the dataset by Region and Income Group, calculate the average, min, and max of GDP per Capita & life Expentancy, total, average, min, and max of population, then save the result to Excel.


```python
df_2018 = df[df["Year"] == 2018]
print(df_2018.shape)
df_2018.sample(5)
```

```python
df_group = df_2018.groupby(["Region", "Income Group"]).aggregate(
    {
     'Life expectancy at birth, total (years)': ['min', 'mean', 'max'],
     'GDP per capita (current US$)': ['min', 'mean', 'max'],
     'Population, total': ['min', 'mean', 'max', 'sum']
    }
)

df_group
```

```python
df_group.to_excel("by_region_income_group.xlsx")
```
